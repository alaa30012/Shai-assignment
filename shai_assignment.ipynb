```python
df = df.drop(df[(df['Id'] >= 148627) & (df['Id'] <= 148653)].index)
# drop rows 148627 to 148653 because as they do not contain useful information
#Dropping rows removes null values in both "overtimepay"and"otherpay,"indicating they no longer contain missing data.
```

```python
df['Benefits'] = df['Benefits'].fillna(df['TotalPayBenefits'] - df['TotalPay'])
#Replace null values in the "benefits" column with the result of subtracting "totalpaybenefits" from "totalpaybenefits," as this equation is used to calculate benefits.
```

```python
df['BasePay'] = df['BasePay'].fillna(df['TotalPay'] - (df['OtherPay'] + df['OvertimePay']))
#fill the null values in 'BasePay' by subtracts the sum of 'OtherPay' and 'OvertimePay' from 'TotalPay'.
```

```python
df = df.drop(['Notes','Status'], axis=1)
#Drop both the 'Notes' and 'Agency' columns as they exclusively contain null values.
```

```python
df.isnull().sum()
#ensure that all data is null
```

```python
import matplotlib.pyplot as plt
plt.hist(df.TotalPay)
plt.title('Distribution of Salaries')
```

```python
df['JobTitle'].value_counts().plot.pie(autopct='%0.2f%%')
```

```python
grouped_data = df.groupby(['JobTitle', 'Year']).agg({
    'TotalPay': 'mean',
    'OtherPay': 'mean',
    'Benefits': 'mean',
    'TotalPayBenefits': 'mean'
}).reset_index()
grouped_data
```

```python
plt.bar(grouped_data['JobTitle'], grouped_data['TotalPay'], color='blue')
plt.xlabel('Job Title')
plt.ylabel('Average Total Pay')
plt.title('Average Total Pay Across Different Job Titles')
plt.show()

```

```python
correlation = df['TotalPay'].corr(df['OtherPay'])
correlation
```

```python
plt.scatter(df['TotalPay'],df['OtherPay'])
plt.title('Scatter Plot - TotalPay vs OtherPay')
plt.xlabel('TotalPay')
plt.ylabel('OtherPay')
```
